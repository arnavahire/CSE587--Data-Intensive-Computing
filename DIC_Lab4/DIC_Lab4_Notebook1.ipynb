hashtag.regex -- contains hashtags derived using regex() function.

str_extract_all() -- extracts all the hashtags from the tweet_text field


```r
install.packages("stringr",repos="http://cran.us.r-project.org/")
```

```r
library(stringr)  
# to extract hashtags
hashtag.regex <- regex("(?<=^|\\s)#\\S+")
hashtags <- str_extract_all(tweetsDF$text, hashtag.regex)
hashtags
hashtags[sapply(hashtags, is.null)] <- NULL  # Removes null records from the hashtag field
```
