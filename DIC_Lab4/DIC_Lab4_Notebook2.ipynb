##### Pre-processing Tweet's 'Text' field

```r
library(stringr)
```

##### Removing emoticons

```r
tweet_textfield <- sapply(tweet_textfield ,function(row) iconv(row, "latin1", "ASCII", sub=""))
```

##### GSUB is used to select a 'pattern' in a given text field and substitute that pattern by the required String.

```r
  tweet_textfield = gsub("&amp", "", tweet_textfield)
  tweet_textfield = gsub("(RT|via)((?:\\b\\W*@\\w+)+)", "", tweet_textfield)
  tweet_textfield = gsub("@\\w+", "", tweet_textfield)
  tweet_textfield = gsub("[[:punct:]]", "", tweet_textfield)
  tweet_textfield = gsub("[[:digit:]]", "", tweet_textfield)
  tweet_textfield = gsub("http\\w+", "", tweet_textfield)
  tweet_textfield = gsub("[ \t]{2,}", "", tweet_textfield)
  tweet_textfield = gsub("^\\s+|\\s+$", "", tweet_textfield) 
```

##### str_replace_all() is also used to filter the data for a given observed pattern. We filter the hashtags , retweets and URLS if any remaining

```r
tweet_textfield <- str_replace_all(tweet_textfield," "," ")
# Get rid of URLs
tweet_textfield <- str_replace_all(tweet_textfield, "http://[a-z,A-Z,0-9]*"," ")
# Take out retweet header, there is only one
tweet_textfield <- str_replace(tweet_textfield,"RT @[a-z,A-Z]*: "," ") 
# Get rid of hashtags
tweet_textfield <- str_replace_all(tweet_textfield,"#[a-z,A-Z]*"," ")
# Get rid of references to other screennames
tweet_textfield <- str_replace_all(tweet_textfield,"@[a-z,A-Z]*"," ")   
```
