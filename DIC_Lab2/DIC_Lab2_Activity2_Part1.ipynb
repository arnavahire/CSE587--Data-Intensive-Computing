###### Finding 'gk_rating':

```r
# Obtaining the 'gk_rating' attribute

myplayer=tbl(my_db,sql("Select distinct Player.player_name,Player.player_api_id,Player_Attributes.overall_rating,Player_Attributes.gk_diving,Player_Attributes.gk_kicking,Player_Attributes.gk_handling,Player_Attributes.gk_positioning,gk_reflexes from Player LEFT JOIN Player_Attributes on Player.player_api_id=Player_Attributes.player_api_id"))
myplayerdf=data.frame(myplayer)
myplayerdf
```

#### 2. Creating 'Team.csv'

```r
# We use Team and Team_Attributes table to extract info for Team.csv

team<-tbl(my_db,sql("Select T.team_long_name,T.team_api_id ,TA.buildUPPlaySpeed,TA.buildUpPlayDribbling,TA.buildUpPlayPassing,TA.chanceCreationPassing,TA.chanceCreationCrossing,TA.chanceCreationShooting,TA.defencePressure,TA.defenceAggression,TA.defenceTeamWidth
from Team T  
left join  Team_Attributes TA 
on  T.team_api_id=TA.team_api_id"))
teamdf=data.frame(team)
teamdf
```

###### Finding 'buildUpPlay', 'chanceCreation' and 'defense':

```r
my_team=mutate(team,buildUpPlay=buildUpPlaySpeed+buildUpPlayDribbling+buildUpPlayPassing)
my_team=mutate(my_team,chanceCreation=chanceCreationPassing+chanceCreationCrossing+chanceCreationShooting)
my_team=mutate(my_team,defense=defencePressure+defenceAggression+defenceTeamWidth)
my_teamdf=data.frame(my_team)
my_teamdf
```

```r
# We group this data by team_long_name

mtdf=sqldf("select * from my_teamdf group by team_long_name")
mtdf
```

##### Finding 'Goals':

```r
m3=tbl(my_db,sql("Select * from Match"))
```

```r
# We collect the team_api_ids for home and away in hgoal and agoal

hgoal <- collect(group_by(m3, home_team_api_id)) 
agoal <- collect(group_by(m3, away_team_api_id)) 
```

```r
# If the team is home team, we select home_team_goal

hgl1 <- select(hgoal,team_api_id=home_team_api_id,goals=home_team_goal)
hddata=data.frame(hgl1)

# We sum all the goals for home team

yo=sqldf("Select team_api_id,SUM(goals) as Goals1 from hddata GROUP BY team_api_id")
#yo

# If team is away team, we select away_team_goal

agl1 <- select(agoal,team_api_id=away_team_api_id,goals=away_team_goal)
addata=data.frame(agl1)

# We sum all the goals for away team

yo1=sqldf("Select team_api_id,SUM(goals) as Goals2 from addata GROUP BY team_api_id")
#yo1

# We join the data frames on team_api_id

yo2=sqldf("Select yo.team_api_id,yo.Goals1,yo1.Goals2 from yo LEFT JOIN yo1 on yo.team_api_id=yo1.team_api_id")
#yo2

# We now add the home goals column and away goals column to obtain final Goals column

yo3=sqldf("Select team_api_id,(Goals1+Goals2) as Goals from yo2")
yo3
```

```r
final_team=merge(mtdf,yo3,by="team_api_id")
#final_team

#select only required columns from the final_team data frame

ffinaltm=select(final_team,team_long_name,buildUpPlay,chanceCreation,defense,Goals)
ffinaltm
write.csv(file="team.csv",ffinaltm)
```
