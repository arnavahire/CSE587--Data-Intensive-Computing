```r
#For reading xls file we install package 'gdata'
install.packages("gdata")
```

```r
require(gdata)
df = read.xls ("NHLTop100.xlsx", sheet = 1, header = TRUE) 
df
```

##### Part a:

```r
linear_model1<-lm(A~G, data=df)
#Initial R-square and P
summary(linear_model1)
```

```r
library(ggplot2)
ggplot(df,aes(x=G,y=A))+xlab("Goals")+ylab("Assists")+
geom_point(aes(col=GP,size=G))+
geom_abline(intercept=494.5135,slope=0.5202,col="red")+ ylim(0,2000) + xlim(0,1000)+
scale_color_gradient2(name="",
                         breaks = c(750,1250,1750),
                         labels = c("750","1250","1750"),
                         low = "red", high = "blue", mid="white", midpoint=1250)
```

##### Initial R-square and P
(Adjusted) R-square: 0.09913 
P value: 0.0008317

##### Part b:

```r
linear_model2<-lm(I(df$A-1963) ~ I(df$G-894))
summary(linear_model2)
```

```r
library(ggplot2)
ggplot(df,aes(x=G,y=A))+xlab("Goals")+ylab("Assists")+
geom_point(aes(col=GP,size=G))+
geom_abline(intercept=-1003.4220 ,slope=3.318145414,col="blue")+ ylim(0,2000) + xlim(0,1000)+
geom_text(data=df[1,],label="Wayne Gretzky")+
scale_color_gradient2(name="",
                         breaks = c(750,1250,1750),
                         labels = c("750","1250","1750"),
                         low = "red", high = "blue", mid="white", midpoint=1250)
```

##### R-square and P for line passing through Wayne Gretzky
(Adjusted) R-square: 0.9082
P value: 2.2e-16
