###### Part e: Interpretation of data

1. 'linear_model1' and 'linear_modeld1' are the linear models for the original regression line. Comparing the R-square and P-Values of both of them, we can say that the the linear regression suffers because of the new outlier point introduced(Patrick Kane). Th R-squared value represents the error and it is low initially, but later it gets increased when the line is made to pass through origin, which is undesirable.
2. The R-squared value is highest for 'linear_model2' when the line is made to pass through 'Wayne Gretzky' since it is an outlier point. Thus more error is introduced because of considering that point in linear regression.
3. The R-Squared value decreases for 'linear_model3' to almost half of the 'linear_model2' on introducing a new point (Patrick Kane) that is closer to the non-outlier points and on the opposite side of the farthest outlier (Wayne Gretzky).
4. The P-Value is highest for the 'linear_model1' which means the predictor is not a good addition to our model. Also it is almost the same for all other models which is significantly less than the 1st model.
